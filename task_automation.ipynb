```python
!pip install yfinance

import yfinance

ticker = input('Action code:')
dados = yfinance.Ticker(ticker)

dados.history()

table = dados.history("6mo")
table

#Select Close

close = table.Close
close

#Line graph
close.plot()

#Statistics
maximum = round(close.max(), 2)
minimum = round (close.min(),2)
nowadays = round (close[-1], 2)

print(maximum)
print(minimum)
print(nowadays)



```
